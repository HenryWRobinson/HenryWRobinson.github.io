```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from zipfile import ZipFile 
from io import BytesIO
import re
import urllib.request
from pathlib import Path
from google.colab import files
```

```python
parser = 'html.parser'
resp = urllib.request.urlopen("https://www.nhsbsa.nhs.uk/prescription-data/dispensing-data/prescription-cost-analysis-pca-data")
soup = BeautifulSoup(resp, parser, from_encoding=resp.info().get_param('charset'))
boop = []
for link in soup.find_all(lambda tag:tag.name=="a" and "PCA data" in tag.text):
  boop.append(link['href'])
egg1 = "https://www.nhsbsa.nhs.uk"
egg2 = "/sites/default/files/"
egg3 = " "
boop = [elem.replace(egg1, '') for elem in boop]
boop = [elem.replace(egg2, '') for elem in boop]
boop = [elem.replace(egg3, '') for elem in boop]
print(boop)


url_base = 'https://www.nhsbsa.nhs.uk/sites/default/files/{}'

file_base ='nhs_{}'
print(file_base)
shrek ="/"

help = ['2021-03/PCA%20Jan21.xlsx', '2021-04/PCA%20Feb21.xlsx', '2021-06/PCA%20Mar21.xlsx','2020-06/PCA%20Apr20.xlsx', '2020-07/PCA%20May20.xlsx', '2020-08/PCA%20Jun20.xlsx', '2020-09/PCA%20Jul20.xlsx', '2020-10/PCA%20Aug20.xlsx', '2020-11/PCA%20Sep20.xlsx', '2020-12/PCA%20Oct20.xlsx', '2021-01/PCA%20Nov20.xlsx', '2021-03/PCA%20Dec20.xlsx']
#test = ['2021-04/PCA%20Feb21.xlsx','2021-06/PCA%20Mar21.xlsx']

for i in boop:
  print(i)
  url = url_base.format(i)
  print(url)
  df = pd.read_excel(url)
  if i not in help:
    new_header = df.iloc[0] 
    df = df[1:] 
    df.columns = new_header
  if i in help:
    new_header = df.iloc[1] 
    df = df[2:] 
    df.columns = new_header
  file_name = file_base.format(i)
  file_name = [elem.replace(shrek, '')for elem in file_name]
  print(df)
  df.to_csv('{}.csv'.format(file_name))
  #files.download('{}.csv'.format(file_name))
  print(file_name)
```
