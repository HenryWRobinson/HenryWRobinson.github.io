```python
import requests
import pandas as pd
import json
from google.colab import files
from pandas import json_normalize
```

```python
url = 'https://api.polygon.io/v2/aggs/ticker/EZJ/range/1/day/2021-01-31/2022-01-07?adjusted=true&sort=asc&limit=362&apiKey=KQ5KSHHKPHpgCHeH3WxK_BQFm4egaxiZ'
r = requests.get(url)
data = r.json()
print(data)
df = pd.json_normalize(data['results'])
df.to_csv('Week9RegEasyJetStockPrice.csv', encoding='utf-8',index=False)
df

files.download('Week9RegEasyJetStockPrice.csv')
```

```python
url = 'https://api.coronavirus.data.gov.uk/v2/data?areaType=overview&metric=newCasesByPublishDate&format=json'
r = requests.get(url)
data = r.json()
df = pd.json_normalize(data)

df.to_csv('Week9RegCovCasebyPub.csv')
files.download('Week9RegCovCasebyPub.csv')


```
