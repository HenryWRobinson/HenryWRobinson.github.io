```python
import requests
import pandas as pd
import os
import numpy as np
import string
from bs4 import BeautifulSoup
```

```python
url = "https://www.national-lottery.co.uk/results/lotto/draw-history"
html = requests.get(url)
soup = BeautifulSoup(html.content, 'html.parser')
date = soup.find_all("li", class_="table_cell table_cell_1 table_cell_1_width_no_raffle table_cell_first")
date_list1=[i.text.replace(" ","").replace("Date","").replace("\n","").replace("Sat","").replace("Wed","") for i in date]
date_list2=date_list1[1:]
print(date_list2)
pot = soup.find_all('li',class_='table_cell table_cell_2 table_cell_2_width_no_raffle')
pot_list1=[i.text.replace(" ","").replace("Jackpot","").replace("\n","").replace("£","").replace(",","") for i in pot]
pot_list2=pot_list1[1:]
print(pot_list2)
                            
```

```python
df = pd.DataFrame(list(zip(date_list2, pot_list2)))
df.columns = ['Date', 'Jackpot(£)']
df
```

```python
df.to_csv("Week7/ScraperNatLotData.csv")
```
